## Part 3 · Assignment Operators 📥


**3.1** — Start `coins = 100`. Use `+=` to add 50, then `-=` to spend 30, then `*=` to double. Print after each step.

```python
coins = 100
# 👉 YOUR CODE HERE
coins = 100
print(f"Start: {coins}")
coins += 50      # add 50 coins = coins + 50
print(f"After += 50: {coins}")
```

```python
coins = 150
print(f"Start: {coins}")
coins -= 30      # subtract 30 coins = coins - 30
print(f"After -= 30: {coins}")
```

```python
coins = 120
print(f"Start: {coins}")
coins *= 2      # multiply 2 coins = coins * 2
print(f"After *= 2: {coins}")
```
